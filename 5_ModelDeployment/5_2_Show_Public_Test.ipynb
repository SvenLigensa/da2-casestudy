```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_test = np.load('./public_test.npy')
```

```python
def convert_bands_to_RGB(array: np.ndarray):
    array = array[2::-1]
    array = array / array.max(axis=(1, 2), keepdims=True)
    array = np.moveaxis(array, 0, -1)
    return array
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
plt.imshow(convert_bands_to_RGB(X_test))
ax.axis('off')
plt.savefig('Public_Test_GroundTruth.png', bbox_inches='tight')
plt.show()
```
