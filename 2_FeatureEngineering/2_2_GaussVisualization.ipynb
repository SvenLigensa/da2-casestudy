# Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Gauss Visualization

```python
x = np.arange(-2, 3)
y = np.arange(-2, 3)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2) / 2) / (2 * np.pi)
Z /= np.sum(Z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(X.flatten()-0.5, Y.flatten()-0.5, np.zeros_like(Z.flatten()), 1, 1, Z.flatten(), color='#942738')
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_zlim(0, np.max(Z))
ax.set_zticks([])
ax.view_init(elev=20, azim=-60)

plt.savefig("GaussVisualization.png", bbox_inches='tight', pad_inches=0, dpi=500)
plt.show()
```
